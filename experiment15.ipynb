```python
import pandas as pd
import numpy as np

data = {
    'Product':['Laptop', 'Mobile', 'Tablet', 'Headphones', 'Smartwatch', 'Camera'],
    'Price':[55000, 20000, 30000, 2000, 8000, 45000],
    'Units_Sold':[25, 60, 35, 80, 50, 15],
    'Discount': [10,15,12,20,18,8]
}
df= pd.DataFrame(data)
print(df)
```

```python
#1 min-max normalization
#2 min-max normalization for prize

df['Price_minmax'] = (df['Price'] - df['Price'].min()) / (df['Price'].max() - df['Price'].min())
print(df[['Product', 'Price', 'Price_minmax']])
print(df)
```

```python
cols= ['Price', 'Units_Sold', 'Discount']
df_norm= (df[cols]-df[cols].min())/(df[cols].max()-df[cols].min())
print(df_norm)
```

```python
cols= ['Price', 'Units_Sold', 'Discount']
df_norm= (df[cols]-df[cols].min())/(df[cols].max()-df[cols].min())
print(df)
```

```python
#Z-score optimization
df['Price_zscore'] = (df['Price'] - df['Price'].mean()) / df['Price'].std()
print(df)
```

```python
column= ['Price', 'Units_Sold', 'Discount']
df_norm= (df[column]-df[column].mean())/df[column].std()
print(df_norm)
```

```python
# decimal scaling
df['Price_decimal'] = df['Price'] / 100000
print(df[['Product', 'Price', 'Price_decimal']])
```

```python
import pandas as pd
import numpy as np
data = {
'Order_ID':[101,102,103,104,105,106],
'Customer_Gender':['Male','Female','Female','Male','Female','Male'],
'Payment_Method':['UPI','Credit Card','Debit Card','COD','UPI','Debit Card'],
'Product_Category':['Electronics','Clothing','Home','Electronics','Beauty','Electronics'],
'City':['Pune','Mumbai','Delhi','Bangalore','Pune','Hyderabad'],
'Order_Value':[4500,2300,3200,5200,1800,4100]
}
df1= pd.DataFrame(data)
print(df1)
```

```python
from sklearn.preprocessing import LabelEncoder

le= LabelEncoder()
df1['Gender_Label']= le.fit_transform(df1['Customer_Gender'])
print(df1)
```

```python
df1['City_Label'] = le.fit_transform(df1['City'])
print(df1)
```

```python
df1['Payment_Label'] = le.fit_transform(df1['Payment_Method'])
print(df1)
```

```python
df1['Product_Label']= le.fit_transform(df1['Product_Category'])
print(df1)
```

```python
df1['Order_Label'] = le.fit_transform(df1['Order_Value'])
print(df1)
```

```python
df1_encoded= pd.get_dummies(df1, columns=['Customer_Gender', 'Payment_Method', 'Product_Category', 'City'])
print(df1_encoded)
```

```python
df1_dummy = pd.get_dummies(df1, columns=['Payment_Method'], drop_first = True)
print(df1_dummy)
```
